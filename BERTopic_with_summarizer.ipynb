```python
from bertopic import BERTopic
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import pandas as pd
import string

# Download NLTK resources (if not downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Assuming 'data' is your pandas DataFrame and 'column_name' is the name of the column containing text data
column_name = 'Clean-Comment'  # Replace 'text_column' with the actual column name

# Function to preprocess text data
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))

    # Tokenize the text
    words = word_tokenize(text.lower())  # Convert to lowercase

    # Remove punctuation
    words = [word for word in words if word.isalpha()]

    # Remove stopwords
    words = [word for word in words if word not in stop_words]

    # Join the words back to a sentence
    return " ".join(words)

# Preprocess the text data in the specified column
data[column_name] = data[column_name].apply(preprocess_text)

# Extract cleaned text data
cleaned_documents = data[column_name].tolist()

# Initialize BERTopic model with cleaned data
model = BERTopic()
topics, _ = model.fit_transform(cleaned_documents)

# Visualize top keywords for 50 topics
model.visualize_barchart(top_n_topics=50)

```

```python
from bertopic import BERTopic

# Initialize a new BERTopic model with a larger number of topics
new_nr_topics = 50  # Replace with the desired number of topics
new_model = BERTopic(nr_topics=new_nr_topics)

# Fit the new BERTopic model on your text data
topics, _ = new_model.fit_transform(data['Clean-Comment'])

# Get the most frequent topics
most_freq_topics = new_model.get_topic_freq()
print(most_freq_topics)
```

```python
# Fit BERTopic on your text data
topics, _ = model.fit_transform(data['Clean-Comment'])

# Get the most frequent words per topic
most_freq_words_per_topic = {}
unique_topics = set(topics)  # Get unique topics

for topic_id in unique_topics:
    most_freq_words = model.get_topic(topic_id)
    most_freq_words_per_topic[topic_id] = most_freq_words

# Display most frequent words per topic
for topic_id, words in most_freq_words_per_topic.items():
    print(f"Most frequent words for Topic {topic_id}:")
    print(words)
    print("\n")
```

```python
import pandas as pd

# Assuming 'data' is your pandas DataFrame and 'column_list' contains the names of columns with text data
column_list = ['Clean-Comment']  # Replace with your column names

# Calculate average text length for each specified column
average_lengths = {}
for column_name in column_list:
    if data[column_name].dtype == 'object':  # Check if the column contains text data (object type)
        average_lengths[column_name] = data[column_name].apply(lambda x: len(str(x))).mean()

# Display average text length for each specified text-containing column
for column, avg_length in average_lengths.items():
    print(f"Average text length in '{column}': {avg_length:.2f} characters")
```

```python
doc = df[df.Query == 'Pudgy Penguins']['Clean-Comment'].sample(random_state=42)
```

```python
# text wrapping function
def wrap(x):
  return textwrap.fill(x, replace_whitespace = False, fix_sentence_endings = True)
```

```python
print(wrap(doc.iloc[0]))
```

```python
 from transformers import pipeline

# Create a summarization pipeline using BART model
summarizer = pipeline("summarization")

# Assuming 'doc' is your DataFrame column with text data
# Summarize the text document at the specified index
summary = summarizer(doc.iloc[0])

# Display the generated summary
print("Generated Summary:", summary[0]['summary_text'])
```
