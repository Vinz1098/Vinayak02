```python
from bertopic import BERTopic
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import pandas as pd
import matplotlib.pyplot as plt

# Download NLTK resources (if not downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Assuming 'data' is your pandas DataFrame and 'column_name' is the name of the column containing text data
column_name = 'Clean-Comment'  # Replace 'text_column' with the actual column name

# Function to preprocess text data
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))

    # Tokenize the text
    words = word_tokenize(text.lower())  # Convert to lowercase

    # Remove punctuation
    words = [word for word in words if word.isalpha()]

    # Remove stopwords
    words = [word for word in words if word not in stop_words]

    # Join the words back to a sentence
    return " ".join(words)

# Preprocess the text data in the specified column
data[column_name] = data[column_name].apply(preprocess_text)

# Extract cleaned text data
cleaned_documents = data[column_name].tolist()

# Initialize BERTopic model with cleaned data
model = BERTopic()
topics, _ = model.fit_transform(cleaned_documents)

# Retrieve and visualize top 10 words for each topic

top_n_topics = 50  # Number of topics to visualize
top_n_words = 10  # Number of top words to display for each topic

topic_freq = model.get_topic_freq()

for topic_id in range(top_n_topics):
    # Check if the topic_id exists in the topic_freq index and if the type is not an integer
    if topic_id in topic_freq.index and not isinstance(topic_freq.loc[topic_id], int):
        top_words = model.get_topic(topic_id)
        words, _ = zip(*top_words)
        words = list(words)[:top_n_words][::-1]

        frequencies = topic_freq.loc[topic_id].head(top_n_words)['Count']

        if isinstance(frequencies, pd.Series):
            frequencies = frequencies.values[::-1]  # Convert to numpy array if it's a Series

        plt.figure(figsize=(10, 6))
        plt.barh(words, frequencies, color='skyblue')
        plt.xlabel('Frequency')
        plt.title(f'Top {top_n_words} Frequent Words for Topic {topic_id}')
        plt.gca().invert_yaxis()
        plt.show()
```

```python
import pandas as pd

# Adjust the display settings to show more characters
pd.set_option('display.max_colwidth', None)

# Initialize an empty list to store top words data
top_words_data = []

# Assuming 'model' and 'topic_freq' are already defined from your previous code
# Assuming 'top_n_topics' and 'top_n_words' are defined

top_n_topics = 50  # Number of topics to visualize
top_n_words = 10  # Number of top words to display for each topic

for topic_id in range(top_n_topics):
    if topic_id in topic_freq.index and not isinstance(topic_freq.loc[topic_id], int):
        top_words = model.get_topic(topic_id)

        # Ensure that top_words is a list of tuples before iterating
        if isinstance(top_words, list) and all(isinstance(item, tuple) for item in top_words):
            words_list = [word for word, _ in top_words]
            top_words_data.append((f"Topic {topic_id}", ", ".join(words_list[:top_n_words])))
        else:
            top_words_data.append((f"Topic {topic_id}", ""))  # Add an empty string for unexpected data types

# Create a DataFrame with "Index Topic" and "10 Freq Words" columns
top_words_df = pd.DataFrame(top_words_data, columns=["Index Topic", "10 Freq Words"])

# Print the DataFrame
print(top_words_df)
```

```python
import pandas as pd

# Assuming 'data' DataFrame contains 'Clean-Comment' column
# Assuming 'model.fit_transform(data['Clean-Comment'])' is executed and 'topics', 'topic_freq' are obtained

# Assuming 'top_n_topics' and 'top_n_words' are defined
top_n_topics = 50  # Number of topics to visualize
top_n_words = 10  # Number of top words to display for each topic

# Initialize an empty list to store top comments data
top_comments_data = []

# Check if 'topics' is a valid array
if isinstance(topics, list):
    for topic_id in range(top_n_topics):
        if topic_id in topic_freq.index and not isinstance(topic_freq.loc[topic_id], int):
            # Filter rows based on the condition (topics == topic_id)
            selected_comments = data[[topic == topic_id for topic in topics]]['Clean-Comment'].head(10).tolist()

            # Combine the comments into a single string
            comments_text = "\n".join(selected_comments)

            top_comments_data.append(comments_text)
        else:
            top_comments_data.append("")  # Add an empty string if topic_id is not in topic_freq index
else:
    print("'topics' variable is not in the expected format. Check the content or how it's generated.")

# Create a DataFrame with "Top 10 Comment" column
top_comments_df = pd.DataFrame(top_comments_data, columns=["Top 10 Comment"])

# Assuming you already have a DataFrame 'top_words_df' created from previous code

# Merge 'top_words_df' and 'top_comments_df' based on the index of 'top_words_df'
final_df = pd.concat([top_words_df, top_comments_df], axis=1)

# Remove rows where both 'Index Topic' and '10 Freq Words' columns are NaN
final_df = final_df.dropna(subset=['Index Topic', '10 Freq Words'], how='all')

# Print the updated DataFrame without rows containing NaN values in 'Index Topic' and '10 Freq Words'
print(final_df)
```
