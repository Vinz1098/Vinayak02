```python
from bertopic import BERTopic

# Assuming 'documents' contain your text data
# Initialize BERTopic
model = BERTopic()

# Fit BERTopic on the documents
topics, _ = model.fit_transform(documents)

# Get topics and their counts
topic_counts = model.get_topic_info()

# Display topics and their counts
print(topic_counts)
```

```python
# Visualize the generated topics
model.visualize_topics()
```

```python
# Get the most frequent words per topic
most_freq_words = model.get_topic(10)  # Replace 0 with the topic number you want to explore
print(most_freq_words)
```
